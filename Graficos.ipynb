```python
import pandas as pd #biblioteca de ciencia de dados abertos.tratamentos de dados , análise exploratórias, ML
#2 objetos principais: series e DataFrames
import matplotlib.pyplot as plt
```

```python
# df = dataframe  -> estrutura de dados bidimensional - linhas e colunas
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRAJ61QW8jxX46erVqpUgk04qET8Axk2SecUELC8U1gzzP4llLT9nwONG8n_K4bsQfqIfjerc-GHa5U/pub?gid=1640484243&single=true&output=csv')
display(df)
```

```python
df.groupby('departamento')['departamento'].count()
# groupby('departamento') = agrupa os dados do dataframe com base nas colunas departamentos
# ['departamento'] = selecionar especificamente a coluna departamento dentro dos grupos criados
# .count() = contar o numero de elementos em cada grupo que foi formado
```

```python
df.groupby('departamento')['departamento'].count().rename('Quantidade de funconários')
```

```python
df.groupby('departamento').agg(qtde=('departamento','count')).sort_values(by='qtde', ascending=False)
#agg = agreggate = agrupar
```

```python
df.groupby('departamento').agg(qtde=('departamento','count')).sort_values(by='qtde', ascending=True)
```
