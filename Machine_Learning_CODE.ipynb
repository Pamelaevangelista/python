```python
#bibliotecas
import pandas as pd # manupulaççao e analise de dados
import numpy as np # operações numéricas e ciêntificas
import matplotlib.pyplot as plt #Criação de gráficpos
import seaborn as sns # visualização de dados baseada no matplotlib

from sklearn.linear_model import LinearRegression
#importar o modelo de regressão linear
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
#importa métricas de avaliação de modelos
from sklearn.model_selection import train_test_split
#importa a função que divide o conjunto de dados em treino e teste
from sklearn.preprocessing import StandardScaler
#importaa classe para escalar os dados, garantindo que todas as variáveis
#estejam na mesma classe

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

import scipy.stats as stats
```

```python
'''Imagine que você quer ensinar um computador a prever algo, como o preço de uma
casa com base no tamanho dela. Em vez de dar uma fórmula pronta, você mostra
vários exemplos reais (dados) e o computador aprende sozinho uma forma de fazer
essa previsão.


O que é Regressão Linear?
A regressão linear é um dos algoritmos mais simples e mais importantes do
machine learning. Ela serve para prever um número (um valor contínuo) com base
em um ou mais dados de entrada.
Exemplo:
Você tem um monte de informações como estas:'''
```

```python
''' Se eu tiver uma casa com 75m², qual será o preço dela?

Como funciona a linha?
A linha de regressão linear tem uma fórmula parecida com a de matemática:

y = a * x +  b
preço = a * tamanho + b

x = valor de entrada (ex.: tamanho da casa)
y = valor que queremos prever (ex.: preço da casa)
a = a inclinação da linha ( mostra o quanto o preço aumenta para cada m² a mais)
b = onde a linha cruza o eixo y (preço-base quando o tamanho é zero)

O algoritmo de regressão linear vai encontrar os valores de "a" e "b" que melhor fazem
essa linha passar perto dos pontos dos seus dados.

Como o computador aprende isso?
Você vai mostrar os dados reais para o computador
Começa com um chute qualquer para "a" e "b"
Compara os valores previstos com os valores reais e verifica o erro.
Ajuste de "a" e "b" um pouco para tentar diminuir o erro.
Repetir isso várias vezes até encontrar a melhor linha.
Esse processo se chama treinamento.

Visualmente(no gráfico):
No eixo X, o tamanho da casa
No eixo Y, o preço da casa
Cada casa vai virar um ponto no gráfico.

'''
```

```python
tamanho = [50, 60, 70, 80]
preco = [100000 , 120000 , 140000, 160000]

tamanho = np.array(tamanho).reshape(-1,1)
'''
np.array(tamanho) = transforma a lista em array numPy, que é o formato que os modelos aceitam
.reshape(-1, 1) = transforma o array em um formato de matriz com uma coluna,
porque o modelo espera esse formato: uma matriz de entrada.
O parâmetro (-1,1) serve para dizer ao numpy: descubra sozinho quantas linhas eu preciso, desde que
tenha uma coluna
'''
modelo = LinearRegression()
modelo.fit(tamanho, preco)
'''
LinearRegression() = cria um modelo de regressão linear (uma reta)
.fit(tamanho, preco) = o modelo aprende a relação entre o tamanho da casa e o preço
'''
tamanho_para_plotar = np.linspace(tamanho.min(), tamanho.max(), 100).reshape(-1, 1)
preco_previsto = modelo.predict(tamanho_para_plotar)

'''
np.linspace(...): cria 100 valores entre o menor e o maior tamanho da casa. Serve para desenhar a linha do gráfico
(regressão) com maior suavidade

modelo.predict(...): o modelo prevê os preços para esses 100 tamanhos - gera pontos da linha vermelha no gráfico
'''

#Plotar
plt.figure(figsize=(10,6))
plt.scatter(tamanho, preco, color='blue', label='Casas (dados reais)' ) #gráfico de dispersão
plt.plot(tamanho_para_plotar, preco_previsto, color='red', linewidth=2, label='Linha de regressão')
plt.xlabel('Tamanho da casa (m²)')
plt.ylabel('Preço da casa (R$)')
plt.title('Regressão Linear: Preço vs. Tamanho da Casa')
plt.legend()
plt.grid(True)
plt.show()
```

Treinando modelo de Regresão linea

1. Preparar os dados
2. Converti em array
3. Formatar para ter uma coluna
4. Instanciei o modelo - iniciada LinearRegression()
5. Treinei o modelo - modelo.fit(tamanho,preco)
6. Predição de modelo - prever os preços para novos tamanhos de casas
7. Visualização - Gráficos

```python
r2 =modelo.score(tamanho, preco) #coeficiente de determinação (R²) -> mede a fração da variância nos dados
print(f'R²: {r2}')
#significa que o modelo explica perfeitamente os dados,
# r²= proximo de zero, não significa nada
```

```python
#Regressão Linear Simples vs Regressão Múltipla
#RLM: Modela a relação entre uma variável dependente e duas ou mais variáveis independentes

'''
Variaável independente: é aquela é manipulação ou controada para avaliar o seu
efeito sobre outra variáel. Ela é independente porque não é afetada por outras
variáveis.

variável dependente: é aquela que é medida ou observada para valiar como ela
muda em resposta às alterações na variável independente
'''
```

```python
'''
Comparando uma base com dados sobre investimento em propaganda e vendas
repita o processo feito anteriormente

propaganda = [1, 2, 3, 4, 5]
vendas = [1, 2, 4, 5, 8]
'''
propaganda = [1, 2, 3, 4, 5]
vendas = [1, 2, 4, 5, 8]

#treinar o modelo

propaganda = np.array(propaganda).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(propaganda, vendas)

tamanho_para_plotar2 = np.linspace(propaganda.min(), propaganda.max(), 100).reshape(-1,1)
vendas_previstas =  modelo.predict(tamanho_para_plotar2)

#plotando
plt.figure(figsize=(10, 6))
plt.scatter(propaganda, vendas, color='blue', label='case vendas(dados reais)')
plt.plot(tamanho_para_plotar2, vendas_previstas, color='red', linewidth=2, label='linha de regressão')
plt.xlabel('propaganda(R$)')
plt.ylabel=('vendas(R$)')
plt.title('Regressão Lineat: Vendas vs Propaganda')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#coeficiente de resposta de modelo
'''
y = a * x + b
vendas = 1.7 * propaganda + b
'''

print(modelo.coef_)#retorna o coeficiente a
print(modelo.intercept_) #retorna o coeficiente b

r2_modelo2 =modelo.score(propaganda, vendas) #coeficiente de determinação (R²) -> mede a fração da variância nos dados
print(f'R²: {r2_modelo2}')
```

```python
'''
Quando estamos tentando estimar y a partir de mais de uma variável independente,
estamos lidando com o que chamamos de Regressão Linear Múltipla

Vamos entender passo a passo:
Na regressão linear simples, temos uma variável independente
(ex: o tamanho da casa).

Fórmula da regressão linear simples:
y = a * x + b

Mas na regressão linear múltipla, temos duas ou mais variáveis independentes
(ex: tamanho da casa e número de quartos e localização).

Exemplo prático:
Vamos imaginar uma planilha com dados sobre casas:
'''
```

image.png

```python
'''Vamos prever o preço da casa (y) com base em:
* tamanho da casa
* número de quartos
* distancia do centro da cidade

Fórmula:
Parecida com a linear simples, mas com mais variáveis:
y = a1 * x1 + a2*x2 + a3*x3...

*x1, x2 .... são variáveis independentes
* a1, a2 .... são pesos ou coeficientes aprendidos pelo modelo
* b é o intercepto (quando vale y quando todas as variáveis forem zero)

Preço = a1* tamanho + a2 * Quartos + a3 * Distancia + b

Visualmente falando...
Com uma variável: a relação é uma linha (2D).

Com duas variáveis: a relação vira um plano (3D).

Com mais de duas: vira um hiperplano
(difícil de imaginar, mas o conceito é o mesmo).

Exemplo simples em linguagem natural:
Se o preço da casa depende de tamanho, número de quartos e distância, então o
modelo vai aprender quanto cada fator influencia no preço.

Exemplo:
Preço = 1500 * tamanho + 20000 *quartos - 10000 * distancia + 50000
cada metro quadrado MULTIPLICA R$ 1500

cada quarto adiciona R$ 20000

Cada km de distância tira R$ 10.000
O valor base (intercepto) é R$ 50.000
'''
```

Cuidados essenciais ao usar Regressão Linear

```python
'''
1. Relação linear entre as variáveis
A regressão linear só funciona bem quando há uma relação linear entre as
variáveis.

✅ O que é isso?
Se você fizer um gráfico da variável x com a y, os pontos devem formar mais ou
menos uma linha reta (no caso simples), ou um padrão que possa ser ajustado por
um plano (no caso múltiplo).

❌ Se a relação for curva (ex: parabólica, exponencial), a regressão linear vai
errar muito.
👉 Dica: Use gráficos de dispersão ou matriz de correlação para verificar se a
relação é linear.
'''
```
